```python
LALschedule.head()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, sharey=True, figsize=(10,5))

sns.histplot(data=LALschedule, x='VISITOR_PTS', kde=True, bins=10, ax=ax1)
sns.histplot(data=LALschedule, x='HOME_PTS', kde=True, bins=10, ax=ax2)
```

```python
home_points = LALschedule[LALschedule['HOME']=='Los Angeles Lakers'][['DATE','HOME_PTS']]
home_points.columns = ['DATE', 'POINTS']

visitors_points = LALschedule[LALschedule['VISITOR']=='Los Angeles Lakers'][['DATE','VISITOR_PTS']]
visitors_points.columns = ['DATE', 'POINTS']

points = pd.concat([home_points, visitors_points])
```

```python
points.sort_index(inplace=True)
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=points, x='DATE', y='POINTS')
```

```python
points['5day_rolling_avg'] = points.POINTS.rolling(5).mean()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=points, x='DATE', y='5day_rolling_avg')
```
