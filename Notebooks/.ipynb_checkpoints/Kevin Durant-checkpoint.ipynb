```python
import matplotlib.pyplot as plt
import seaborn as sns
%config Completer.use_jedi = False
%matplotlib inline
```

```python
import pandas as pd
from datetime import timedelta
import datetime
```

```python
kd2022 = pd.read_csv('kd2022.csv')
```

```python
kd2022.head()
```

## Some simple EDA

```python
sns.set(rc={'figure.figsize':(10,5)})
```

```python
kd2022['DISTANCE'] = kd2022['DISTANCE'].apply(lambda d: d[:-2])
```

```python
kd2022['DISTANCE'] = kd2022['DISTANCE'].apply(lambda d: int(d))
```

```python
# plt.xticks(rotation=45)
sns.histplot(data=kd2022, x='DISTANCE', kde=True, hue='MAKE_MISS')
```

```python
def seconder(x):
    t = pd.to_datetime("00:"+x[:-2])
    td = timedelta(minutes=t.minute, seconds=t.second)
    return td.total_seconds()
```

```python
kd2022['SECONDS_REMAINING'] = kd2022['TIME_REMAINING'].apply(seconder)
```

```python
sns.scatterplot(data=kd2022, x="SECONDS_REMAINING", y='DISTANCE', hue='MAKE_MISS')
```

```python
sns.countplot(data=kd2022, hue='MAKE_MISS', x='QUARTER')
```
